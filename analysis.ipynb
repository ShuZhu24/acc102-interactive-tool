```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\67353\Desktop\acc102-project\date\CocaCola_Ventas_Mundiales.csv')
df.head()
df.info()
df.describe()
```

```python
df.columns
```

```python
df.info()
df.head()
```

```python

print(f"总销售额: ${(df['Quantity'] * df['Unit Price (USD)']).sum():,.0f}")

country_sales = df.groupby('País')['Quantity'].sum().sort_values(ascending=False)
print("\n国家销量排行:")
print(country_sales)

channel_sales = df.groupby('Canal de Venta')['Quantity'].sum().sort_values(ascending=False)
print("\n销售渠道排行:")
print(channel_sales)

monthly_sales = df.groupby('Mon')['Quantity'].sum()
print("\n月度销量趋势:")
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
country_sales.head(8).plot(kind='bar')
plt.title('Top 8 Countries by Sales Volume') 
plt.xlabel('Country')  
plt.ylabel('Sales Volume') 
plt.tight_layout()
plt.savefig('country_sales.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans' 
plt.rcParams['axes.unicode_minus'] = False 
plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')     
plt.xlabel('Month')               
plt.ylabel('Sales Quantity')       
plt.grid(True)

plt.savefig('monthly_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Revenue'] = df['Quantity'] * df['Unit Price (USD)']

df['Total_Cost'] = df['Quantity'] * df['Unit Cost (USD)']

df['Profit'] = df['Revenue'] - df['Total_Cost']

yearly_profit = df.groupby('Year')['Profit'].sum()

plt.rcParams['font.family'] = 'DejaVu Sans'
plt.figure(figsize=(8, 8))
plt.pie(yearly_profit, labels=yearly_profit.index, autopct='%1.1f%%')
plt.title('Profit Distribution by Year')  
plt.savefig('yearly_profit.png', dpi=300)
plt.show()

```
